```python
import pandas as pd
```

```python
data = pd.read_csv('../data/housing.csv')
```

```python
import matplotlib.pyplot as plt

for column in data.columns:
    plt.figure(figsize=(6, 4))
    plt.hist(data[column])
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```
